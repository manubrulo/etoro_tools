```python

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_theme()
from tools import plot_tools, data_utils

import matplotlib.style as style
style.use('seaborn-colorblind')

from rich.console import Console
from rich.table import Table

```

```python

etoroExcel = pd.ExcelFile("etoro-account-statement.xlsx")

df_closed_positions = data_utils.process_df(etoroExcel.parse('Closed Positions'))

```

```python
plot_tools.plot_profit(df_closed_positions, period = 10, colname="Profit Euro")
```

```python
table = Table(title="Action counts")

table.add_column("Type", justify="right", style="cyan")
table.add_column("Name", style="white", no_wrap=True)
table.add_column("Count", justify="right", style="green")
table.add_column("Profit Euro", justify="right", style="magenta")

for action_type in df_closed_positions.groupby('Action'):
    table.add_row(action_type[0].split(' ')[0], action_type[0].split(' ',1)[1], str(action_type[1]['Action'].count()), str(action_type[1]['Profit Euro'].sum().round(2)))

console = Console()
console.print(table)
```

```python
plt.figure(figsize=(30,50))
fig = df_closed_positions[df_closed_positions['Action'].str.contains('Buy')].groupby("Action")['Profit Euro'].sum().sort_values().plot.bar(backend='plotly')
# fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide', autosize=False,
    width=1500,
    height=800,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),)
# fig.update_xaxes(showticklabels=False)
fig.show()
```
